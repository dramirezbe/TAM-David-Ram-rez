```python
from sklearn import datasets

faces = datasets.fetch_olivetti_faces()
X = faces.images
```

```python
import matplotlib.pyplot as plt

print(X.shape)
plt.imshow(X[0],cmap = 'gray')
plt.show()


X = X.reshape(X.shape[0],-1) #Convierte la matriz en un vector gigante
print(X.shape)
```

```python
import numpy as np

U,S,Vh = np.linalg.svd(X)  #Descomposici√≥n de valor singular

print(U.shape, S.shape, Vh.shape)
```

```python
Xe = U.dot(np.diag(S)).dot(Vh[:S.shape[0]])
print(Xe.shape)
```

```python
# Norma de Frobenius, norma multidimensional
error = np.linalg.norm(X-Xe,ord='fro')

print(error)
```
