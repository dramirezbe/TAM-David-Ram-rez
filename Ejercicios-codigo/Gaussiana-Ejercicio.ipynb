```python
import matplotlib.pyplot as plt
import numpy as np



def graficador(s,n): #media, varianza, muestreo

    m = 0.5
    x1 = np.random.normal(m,s,n)
    x2 = np.random.normal(m+1,s,n)

    y1 = np.random.normal(m-0.25,s-0.3,n)
    y2 = np.random.normal(m+0.25,s-0.3,n)

    fig, axs = plt.subplots(2, 2, figsize=(10, 8))

    axs[0,0].scatter(x1,y1)
    axs[0,0].scatter(x2,y2)

    axs[0,0].set_xlim(0,2)
    axs[0,0].set_ylim(0,1)

    """Histograma"""

    yt = np.concatenate((y1,y2))

    county = 0
    for i in range(len(yt)):
        if yt[i] <= 0.5:
            county += 1

    py = county/len(yt)
    p_arr = [py,1-py]

    print(p_arr)

    axs[0,1].barh(["$Y=1$", "$Y=2$"],p_arr)

    axs[1,0].hist([np.concatenate((x1,x2))],bins = 30)
    x_gauss = np.linspace(-1.5, 2.5, 100)
    y_gauss = (1/(np.sqrt(2*np.pi*s)))*(np.e**(-(x_gauss-m)**2/(2*s)))
    axs[1,0].plot(x_gauss,25*y_gauss)
    axs[1,0].set_title("P(x)")

    axs[1,1].hist(x1, bins = 30)
    axs[1,1].set_title("P(x|y=1)")






graficador(0.4,100)






```

$P(x \mid \mu, \sigma^{2}) = \frac{1}{\sqrt{2\pi \sigma^{2}}} exp \left(- \frac{(x - \mu)^{2}}{2\sigma^{2}} \right)$
